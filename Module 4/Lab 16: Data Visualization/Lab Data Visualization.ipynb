**Create a SQLite Database and Insert the Data**

Now, let's create a new SQLite database (`survey-data.sqlite`) and insert the data from the DataFrame into a table using the sqlite3 library:

## Hands-on Lab


### Visualizing the Distribution of Data

**Histograms**

Plot a histogram of CompTotal (Total Compensation).


```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming conn is your open SQLite connection
QUERY = """
SELECT CompTotal FROM main WHERE CompTotal IS NOT NULL
"""

df = pd.read_sql_query(QUERY, conn)

# Plotting histogram directly on the 'CompTotal' column with 50 bins
df['CompTotal'].hist(bins=50, figsize=(10,6), color='skyblue', edgecolor='black')

plt.xlabel('Total Compensation (USD)')
plt.ylabel('Frequency')
plt.title('Histogram of Total Compensation (CompTotal)')
plt.grid(True)
plt.show()

```

**Box Plots**

Plot a box plot of Age.


```python
## Write your code here
QUERY = """ SELECT * FROM main """ 
df = pd.read_sql_query(QUERY, conn)

# Map age ranges to numerical values
age_map = {
    'Under 18 years old': 0,
    '18-24 years old': 1,
    '25-34 years old': 2,
    '35-44 years old': 3,
    '45-54 years old': 4,
    '55-64 years old': 5,
    '65 years or older': 6,
    'Prefer not to say': 7
}

df['AgeNum'] = df['Age'].map(age_map)

# Plot the box plot for AgeNum
boxplot = df.boxplot(column='AgeNum', vert=True, patch_artist=True, boxprops=dict(facecolor='lightblue'))

# Customize y-axis ticks with age categories
plt.yticks(
    ticks=list(age_map.values()),
    labels=list(age_map.keys())
)

plt.xlabel('Age Category Numeric Code')  # Clarify x-axis meaning
plt.ylabel('Age Groups')
plt.title('Box Plot of Age Categories (Numeric Codes)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### Visualizing Relationships in Data

**Scatter Plots**

Create a scatter plot of Age and WorkExp.


```python
## Write your code here
QUERY = """ SELECT * FROM main """ 
df = pd.read_sql_query(QUERY, conn)

# Map age ranges to numerical values
age_map = {
    'Under 18 years old': 0,
    '18-24 years old': 1,
    '25-34 years old': 2,
    '35-44 years old': 3,
    '45-54 years old': 4,
    '55-64 years old': 5,
    '65 years or older': 6,
    'Prefer not to say': 7
}

df['AgeNum'] = df['Age'].map(age_map)

# Create a scatter plot of Age and WorkExp
plt.figure(figsize=(10, 6))
plt.scatter(df['AgeNum'], df['WorkExp'], alpha=0.5)

# Customize x-axis ticks with age categories
plt.xticks(
    ticks=range(len(age_map)),
    labels=[
        'Under 18 years old',
        '18-24 years old',
        '25-34 years old',
        '35-44 years old',
        '45-54 years old',
        '55-64 years old',
        '65 years or older',
        'Prefer not to say'
    ],
    rotation=45
)

plt.xlabel('Age Groups')
plt.ylabel('Work Experience (Years)')
plt.title('Scatter Plot of Age vs Work Experience')
plt.grid(True)
plt.show()
```

**Bubble Plots**

Create a bubble plot of `TimeSearching` and `Frustration` using the Age column as the bubble size.


```python
import matplotlib.pyplot as plt

# Fetch data from the SQLite database
QUERY = """ SELECT * FROM main """
df = pd.read_sql_query(QUERY, conn)

# Map age ranges to numerical values
age_map = {
    'Under 18 years old': 0,
    '18-24 years old': 1,
    '25-34 years old': 2,
    '35-44 years old': 3,
    '45-54 years old': 4,
    '55-64 years old': 5,
    '65 years or older': 6,
    'Prefer not to say': 7
}
df['AgeNum'] = df['Age'].map(age_map)

# Drop rows with missing values in TimeSearching, Frustration, or AgeNum
df_clean = df.dropna(subset=['TimeSearching', 'Frustration', 'AgeNum'])

# Scale AgeNum for bubble sizes (adjust multiplier for visual effect)
bubble_sizes = df_clean['AgeNum'] * 200

plt.figure(figsize=(12, 8))
plt.scatter(
    df_clean['TimeSearching'],
    df_clean['Frustration'],
    s=bubble_sizes,    # Size of bubbles
    alpha=0.5,
    edgecolors='w'
)

plt.xlabel('Time Searching')
plt.ylabel('Frustration')
plt.title('Bubble Plot of Time Searching vs Frustration (Bubble size = Age groups)')
plt.grid(True)
plt.show()
```

### Visualizing Composition of Data

**Pie Charts**

Create a pie chart of the top 5 databases(`DatabaseWantToWorkWith`) that respondents wish to learn next year.


```python
import matplotlib.pyplot as plt

# Fetch data from the SQLite database
QUERY = """ SELECT DatabaseWantToWorkWith FROM main WHERE DatabaseWantToWorkWith IS NOT NULL """
df = pd.read_sql_query(QUERY, conn)

# Split multiple values separated by ';' and explode into separate rows
df_exploded = df['DatabaseWantToWorkWith'].str.split(';').explode()

# Count frequencies of each database and get top 5
top_databases = df_exploded.value_counts().head(5)

# Plot pie chart
plt.figure(figsize=(8, 8))
top_databases.plot.pie(autopct='%1.1f%%', startangle=90, counterclock=False,
                       colors=plt.cm.Paired.colors)
plt.ylabel('')  # Remove y-label for cleaner look
plt.title('Top 5 Databases Respondents Want to Learn Next Year')
plt.show()
```

**Stacked Charts** 

Create a stacked bar chart of median `TimeSearching` and `TimeAnswering` for the age group 30 to 35.


```python
import matplotlib.pyplot as plt

# Function to convert time descriptions into numeric values (in minutes)
def time_to_minutes(time_str):
    if pd.isna(time_str):
        return None
    time_str = str(time_str)
    if 'Less than 15' in time_str:
        return 7.5
    elif '15-30' in time_str:
        return 22.5
    elif '30-60' in time_str:
        return 45
    elif '60+' in time_str:
        return 60
    return None  # If time_str is not recognized or is a different format

# Fetch data from the SQLite database
QUERY = """ SELECT * FROM main """
df = pd.read_sql_query(QUERY, conn)

# Map age ranges to numerical values
age_map = {
    'Under 18 years old': 0,
    '18-24 years old': 1,
    '25-34 years old': 2,
    '35-44 years old': 3,
    '45-54 years old': 4,
    '55-64 years old': 5,
    '65 years or older': 6,
    'Prefer not to say': 7
}
df['AgeNum'] = df['Age'].map(age_map)

# Filter the data for 25-34 years old age group (AgeNum == 2)
df_age_group = df[df['AgeNum'] == 2].copy()

# Convert time descriptions to numeric values
df_age_group['TimeSearchingMinutes'] = df_age_group['TimeSearching'].apply(time_to_minutes)
df_age_group['TimeAnsweringMinutes'] = df_age_group['TimeAnswering'].apply(time_to_minutes)

# Drop rows where both are NaN to avoid skewing median calculations
df_age_group = df_age_group.dropna(subset=['TimeSearchingMinutes', 'TimeAnsweringMinutes'], how='all')

# Calculate median values
median_time_searching = df_age_group['TimeSearchingMinutes'].median()
median_time_answering = df_age_group['TimeAnsweringMinutes'].median()

# Plot stacked bar chart using matplotlib
plt.figure(figsize=(8, 6))
plt.bar('25-34 years', median_time_searching, label='Median Time Searching', color='skyblue')
plt.bar('25-34 years', median_time_answering, bottom=median_time_searching, label='Median Time Answering', color='lightgreen')

plt.ylabel('Median Time (minutes)')
plt.title('Median Time Searching and Time Answering for Age Group 25-34')
plt.legend()
plt.show()
```

### Visualizing Comparison of Data

**Line Chart**

Plot the median `CompTotal` for all ages from 45 to 60.


```python
import matplotlib.pyplot as plt
import pandas as pd

# Fetch data from the SQLite database
QUERY = """ SELECT * FROM main """ 
df = pd.read_sql_query(QUERY, conn)

# Define age groups of interest and map to numeric values
age_map = {
    '45-54 years old': 4,
    '55-64 years old': 5
}
df['AgeNum'] = df['Age'].map(age_map)

# Filter the data for the age groups 45-54 and 55-64
df_age_group = df[df['AgeNum'].isin([4, 5])]

# Calculate the median CompTotal for each age group
median_comp = df_age_group.groupby('Age')['CompTotal'].median().reset_index()

# Create a line chart
plt.figure(figsize=(10, 6))
plt.plot(median_comp['Age'], median_comp['CompTotal'], marker='o', linestyle='-', color='b')

# Customize the plot
plt.xlabel('Age Group', fontsize=12)
plt.ylabel('Median Compensation Total (USD)', fontsize=12)
plt.title('Median CompTotal for Age Groups 45-60', fontsize=16)
plt.grid(True)

# Save the plot to a file
plt.savefig('line_chart_median_comptotal.png')

# Display the plot
plt.show()
```

**Bar Chart**

Create a horizontal bar chart using the `MainBranch` column.


```python
import matplotlib.pyplot as plt

# Fetch data from the SQLite database
QUERY = """ SELECT MainBranch FROM main """
df = pd.read_sql_query(QUERY, conn)

# Count the occurrences of each unique value in the MainBranch column
mainbranch_counts = df['MainBranch'].value_counts()

# Sort counts ascending for a nice horizontal bar chart order
mainbranch_counts = mainbranch_counts.sort_values()

# Create a horizontal bar chart
plt.figure(figsize=(10, 7))
mainbranch_counts.plot(kind='barh', color='skyblue', edgecolor='black')

# Customize the plot
plt.xlabel('Count', fontsize=12)
plt.ylabel('Main Branch', fontsize=12)
plt.title('Distribution of MainBranch', fontsize=16)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()

# Save the plot
plt.savefig('horizontal_bar_chart_mainbranch.png')

# Display the plot
plt.show()
```
